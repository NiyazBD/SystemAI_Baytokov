## Задание 1. Расчет арифметических выражений

### z = 12y - $\dfrac{3x+2a+a}{3a+2x-y}$ + 12*y + $\dfrac{\dfrac{\dfrac{x}{y}+2+\dfrac{a}{y}}{2-a+1}}{x}$

```python
x = int(input("Введите переменную x"))
y = int(input("Введите переменную y"))
a = int(input("Введите переменную a"))

z = 12*y - (3*x+2*a+a)/(3*a+2*x-y) + 12*y + (x/y+2+a/y)/(2-a+1)/x
print("Переменная z =",z)
```


## Задание 2. Расчеты математических выражений

УСЛОВИЕ:
### z = $\dfrac{sin(x)+cos(y)}{cos(x)-sin(y)}*tg(x*y)$
\

### УСЛОВИЕ:

```python
import math as m
print("Введите значение переменных:")
x1 = int(input("Введите начальное значение X1: "))
x2 = int(input("Введите конечное значение X2: "))
s = float(input("Введите шаг S: "))
print('***************************************')
print('Резуьтат:')
print('+----------------+')
print('|  X  |   f(x)   |')
print('+----------------+')
while (x1<=x2):
  y = 2**(-1)+(5*x1**2)/(3*x1**3)-m.cos(x1**2)+m.sin(2*x1)
  print('|{:5.2f}|{:10.2f}|'.format(x1,y))
  x1=x1+s
print('+----------------+')
print('***************************************')
```

## Задание 6. Списки (одномерные массивы) в Python

```python
import numpy as np
N = int(input("Введите количество элементов в массиве: "))
print("*****************************************************")
print("Введите диапазон значений")
while True:
  n1 = int(input("Введите отрицательный элемент массива: "))
  n2 = int(input("Введите положительный элемент массива: "))
  if n1>n2: print("Некорректный ввод. Проверьте условия n1<n2")
  else: break
print("*****************************************************")
array = np.random.randint(n1, n2+1, size=N)
array = np.sort(array)[::-1]
print("Сгенерированный одномерный массив")
с = 0
for i in array:
  с = с+1
  print(с,") ",i)
print("*****************************************************")
while True:
  K = int(input("Введите K: "))
  L = int(input("Введите L: "))
  if K>L: print("Некорректный ввод. Проверьте условия K<L")
  else: break
print("*****************************************************")
selected_elements = array[K-1:L]
mean = np.mean(selected_elements)
print(f"Среднее арифметическое элементов массива с номерами от {K} до {L} включительно: {mean}")
```

##Задание 7. Двумерные массивы (матрицы)

УСЛОВИЕ: Дан целочисленный двумерный массив N на M элементов, заполненный случайными числами в заданном диапазоне (от n1 до n2) по убыванию значений в строках.
Найти среднее арифметическое элементов массива с номерами от K до L включительно в заданном столбце С.
Произвести проверку на корректность ввода интервальных значений (n1<n2, K<L, С и т.д.).


```python
import random
import numpy as np
print("Введите размерность массива:")
while True:
  N = int(input("Количество строк: "))
  M = int(input("Количество столбцов: "))
  if N<=0 or M<=0: print("Некоректный ввод. Значения должны быть больше 0.")
  else: break
print("*****************************************************")
print("Введите диапазон значений:")
while True:
  n1 = int(input("Введите минимальное значение массива: "))
  n2 = int(input("Введите максимальное значение массива: "))
  if n1<n2: break
  else: print("Некорректный ввод. Проверьте условия n1<n2")
print("*****************************************************")
array = [[random.randint(n1, n2) for i in range(M)] for i in range(N)]
for row in array:
        row.sort(reverse=True)
print(f"Созданная матрица размерностью {N} на {M}:")
def print_matrix(matrix):
  for row in matrix:
    for element in row:
        print(f"{element:6}", end=" ")
    print()
print_matrix(array)
print("*****************************************************")
while True:
  C = int(input("Введите номер столбца для нахождения среднего значения: "))
  if C>M or C<=0: print("Некоректный ввод. Значение должно быть 0<C<M.")
  else: break
print("*****************************************************")
while True:
  K= int(input("Введите K: "))
  L = int(input("Введите L: "))
  if K>L: print("Некоректный ввод. Проверьте условия K<L.")
  else: break
print("*****************************************************")


sum = 0
count = 0
for i in range(K, L+1):
    sum += array[i][C]
    count += 1
average = sum / count if count > 0 else 0
print("Среднее значение элементов в данном столбце = ", average)


```
